```python
import ast
#to convert string to list
```

```python
def convert(dict):
    list=[]
    for i in ast.literal_eval(dict):
        list.append(i['name'])
    return list
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert3(dict):
    list=[]
    counter=0
    for i in ast.literal_eval(dict):
        if counter!=3:
            list.append(i['name'])
            counter+=1
        else:
            break
    return list
```

```python
movies['cast'].apply(convert3)
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
def fetch_director(dict):
    list=[]
    for i in ast.literal_eval(dict):
        if i['job']=='Director':
            list.append(i['name'])
            break
    return list
```

```python
movies['crew'].apply(fetch_director)
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'].apply(stem)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
chunk_size = 10000
with open("my_table.csv", "w", encoding="utf-16") as f:
    new_df.head(chunk_size).to_csv(f, index=False, header=True)
    for i in range(1, len(new_df) // chunk_size + 1):
        new_df.iloc[i*chunk_size:(i+1)*chunk_size].to_csv(f, index=False, header=False)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors).shape
```

```python
similarity=cosine_similarity(vectors)
```

```python
sorted(similarity[0],reverse=True)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    if movie not in new_df['title'].values:
        print(f"Sorry, the movie {movie} is not in our database.")
        return
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list= sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print("ID: ",(new_df.iloc[i[0]].movie_id))
        print("Title: ",(new_df.iloc[i[0]].title))
        print("Vote Average: ",(new_df.iloc[i[0]].vote_average))
```

```python
user_movie=input("Enter movie name : ").title()
recommend(user_movie)
```
